```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Processing

```python
df = pd.read_csv('uber-raw-data-apr14.txt', sep=",", header=None)
df.columns = ["Date/Time", "Lat", "Lon", "Base"]
```

```python
df.head()
```

```python
df.shape
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

```python
df.tail()
```

```python
#Extract the day

def day(df):
    return df.day

def week(df):
    return df.weekday()

def hour(df):
    return df.hour

    
df['Day'] = df['Date/Time'].apply(day)
df['Week'] = df['Date/Time'].apply(week)
df['Hour'] = df['Date/Time'].apply(hour)
```

# Data Analysis

```python
plt.figure(figsize=(8,5))
sns.set()
plt.hist(df.Day,bins=30)
plt.xlabel("Date")
plt.ylabel("Frequency Of Rides")
plt.title("Frequency of Rides On Each Day of the month")
```

```python
for day, number in df.groupby('Day'):
    print("Day ",day," has",len(number)," number of rides")
```

```python
def count_rows(rows):
    return len(rows)

by_date = df.groupby('Day').apply(count_rows)
by_date_sorted=by_date.sort_values()
by_date_sorted
```

```python
plt.plot(by_date)
```

```python
plt.figure(figsize=(8,5))
plt.bar(range(1,31),by_date_sorted)
```

```python
plt.figure(figsize=(8,5))
sns.set()
plt.hist(df.Hour,bins=24)
plt.xlabel("Hour")
plt.ylabel("Frequency Of Rides")
plt.title("Frequency of Rides On Each Hour of the Day")
```

```python
plt.figure(figsize=(8,5))
sns.set()
plt.hist(df.Week,bins=7)
plt.xlabel("Hour")
plt.ylabel("Frequency Of Rides")
plt.title("Frequency of Rides On Each Hour of the Day")
```
